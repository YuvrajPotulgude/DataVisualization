**Format Notations**
  * %d => day of month
  * %m => Month
  * %Y => Year with century
  * %y => Year without century
  * %I => Represents time in 24 Hour format
  * %H => Represents time in 12 Hour format
  * %M => Minute
  * %p => AM or PM

```python
def quarter(month):
  if month in [1,2,3]:
    return "Q1"
  elif month in [4,5,6]:
    return "Q2"
  elif month in [7,8,9]:
    return "Q3"
  else:
    return "Q4"
```

```python
df['O_quarter'] = df['O_Month'].apply(quarter)
```
